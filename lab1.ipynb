```python
import pint
from scipy import constants
from matplotlib import pyplot as plt
import numpy as np
from open_atmos_jupyter_utils import show_plot
```

```python
si = pint.UnitRegistry()
```

```python
llllll=(100*si.km+100*si.cm+10*si.foot)/(100*si.s**2)
print(llllll)
print(llllll.to_base_units())
print(llllll.to(si.km/si.s**2))
R=8.31*si.J/(si.mol*si.K)
mmol=28.966*si.kg/si.kmol
print(constants.R)
stosunek=(constants.R*si.J/(si.mol*si.K)/mmol)#.to_base_units()
print(stosunek)
```

```python
def gestos(p,T,stosunek):
    return p/(stosunek*T)
```

```python
gestos(1013*si.hPa,300*si.K,stosunek).to_base_units()

```

```python
si.setup_matplotlib()
temp=np.linspace(280,300,100)*si.K
plt.plot(temp,gestos(1013*si.hPa,temp,stosunek).to_base_units(),label="gestosc powietrza w sali ")
# plt.xlabel("temp [K]")
# plt.ylabel(f"gestosc [kilogram2/(meter3 mole)]")
plt.gca().xaxis.set_units(si.celsius)
plt.grid(True)
plt.legend()
show_plot("gestosc_powietrza")
```
